```python
import re
restrictionSites = {re.compile('CCA......TGG'):1, #BstXI
                   re.compile('GCT.AGC'):1, #BlpI
                   re.compile('CCTGCAGG'):0, #SbfI
                   re.compile('CTCGAG'):0, #XhoI
                   re.compile('CCTAGG'):0, #AvrII
                   re.compile('TCTAGA'):0} #XbaI

def matchREsites(sequence, REdict):
    seq = sequence.upper()
    for resite, numMatchesExpected in restrictionSites.iteritems():
        if len(resite.findall(seq)) != numMatchesExpected:
            return False
        
    return True
```

```python
v1Groups = v1Table.groupby(v1Table[('library table v1','gene')])

upstreamConstant = 'CCACCTTGTTG'
downstreamConstant = 'GTTTAAGAGCTAAGCTG'

doublesSgs = []
for i, gene in enumerate(maxGammas.loc[maxDiscs >= 20].loc[maxGammas.abs() >= .05].index.values):
    geneSgs = []
    
    v1sgs = v1Groups.get_group(gene)['library table v2']
    
    for sgId, libraryRow in pd.concat((v1sgs['sgId_v2'],v1sgs['sequence'],phenotypeTable[('20150118_ctx_rigo_crispra','gamma_day10','ave_Rep2_Rep1')],phenotypeTable[('20150118_ctx_rigo_crispra','gamma_day10','ave_Rep2_Rep1')].abs()), keys=('sgId_v2','protospacer sequence','growth phenotype','abs growth phenotype'),axis=1, join='inner').reset_index().set_index('sgId_v2').sort_values('abs growth phenotype', ascending=False).iterrows():
        if len(geneSgs) < 2 and matchREsites(upstreamConstant + libraryRow['protospacer sequence'] + downstreamConstant,restrictionSites):
            libraryRow['gene'] = gene
            libraryRow['oligo sequence'] = upstreamConstant + libraryRow['protospacer sequence'] + downstreamConstant
            geneSgs.append(libraryRow)

            if len(geneSgs) == 1 and np.sign(geneSgs[0]['growth phenotype']) != np.sign(libraryRow['growth phenotype']):
                print gene

    doublesSgs.extend(geneSgs)
        
doubleSgTable = pd.DataFrame(doublesSgs)
doubleSgTable
```
